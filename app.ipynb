```python
# %% [markdown]
# # 🔍 IP Address Analyzer
# ## Simple Python Program to Analyze IP Addresses

# %%
"""
IP Address Analyzer - Complete Program
Accepts IP address and identifies: validity, class/type, public/private, default subnet
"""

import ipaddress

def analyze_ipv4(ip_str):
    """Analyze IPv4 address: class, public/private, default subnet."""
    ip_obj = ipaddress.IPv4Address(ip_str)
    first_octet = int(ip_str.split('.')[0])
    
    # Determine IP class based on first octet
    if 1 <= first_octet <= 126:
        ip_class = 'A'
        default_subnet = '/8'
        subnet_mask = '255.0.0.0'
    elif 128 <= first_octet <= 191:
        ip_class = 'B'
        default_subnet = '/16'
        subnet_mask = '255.255.0.0'
    elif 192 <= first_octet <= 223:
        ip_class = 'C'
        default_subnet = '/24'
        subnet_mask = '255.255.255.0'
    elif 224 <= first_octet <= 239:
        ip_class = 'D (Multicast)'
        default_subnet = 'N/A'
        subnet_mask = 'N/A'
    elif 240 <= first_octet <= 255:
        ip_class = 'E (Reserved)'
        default_subnet = 'N/A'
        subnet_mask = 'N/A'
    else:
        ip_class = 'Unknown'
        default_subnet = 'N/A'
        subnet_mask = 'N/A'
    
    # Determine if public or private
    if ip_obj.is_private:
        scope = 'Private'
    else:
        scope = 'Public'
    
    # Special case for loopback
    if first_octet == 127:
        scope = 'Private (Loopback)'
    
    return {
        'ip': ip_str,
        'version': 'IPv4',
        'class': ip_class,
        'scope': scope,
        'default_subnet': default_subnet,
        'subnet_mask': subnet_mask
    }

def analyze_ipv6(ip_str):
    """Analyze IPv6 address: type, public/private, default subnet."""
    ip_obj = ipaddress.IPv6Address(ip_str)
    ip_int = int(ip_obj)
    
    # Determine IPv6 type based on prefix ranges
    if 0x20000000000000000000000000000000 <= ip_int <= 0x3FFFFFFFFFFFFFFFFFFFFFFFFFFFFFFF:
        ip_type = 'Global Unicast'
        scope = 'Public'
        default_subnet = '/64'
    elif 0xFC000000000000000000000000000000 <= ip_int <= 0xFDFFFFFFFFFFFFFFFFFFFFFFFFFFFFFF:
        ip_type = 'Unique Local (ULA)'
        scope = 'Private'
        default_subnet = '/64'
    elif 0xFE800000000000000000000000000000 <= ip_int <= 0xFEBFFFFFFFFFFFFFFFFFFFFFFFFFFFFF:
        ip_type = 'Link-Local'
        scope = 'Private'
        default_subnet = 'N/A'
    elif 0xFF000000000000000000000000000000 <= ip_int <= 0xFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFFF:
        ip_type = 'Multicast'
        scope = 'Reserved'
        default_subnet = 'N/A'
    elif ip_obj.is_loopback:
        ip_type = 'Loopback'
        scope = 'Private'
        default_subnet = 'N/A'
    else:
        ip_type = 'Other'
        scope = 'Unknown'
        default_subnet = '/64' if ip_obj.is_global else 'N/A'
    
    return {
        'ip': ip_str,
        'version': 'IPv6',
        'type': ip_type,
        'scope': scope,
        'default_subnet': default_subnet
    }

def analyze_ip(ip_str):
    """Main function to validate and analyze IP address."""
    ip_str = ip_str.strip()
    
    if not ip_str:
        return {'error': 'No IP address provided'}
    
    try:
        ip_obj = ipaddress.ip_address(ip_str)
        
        if isinstance(ip_obj, ipaddress.IPv4Address):
            return analyze_ipv4(ip_str)
        else:
            return analyze_ipv6(ip_str)
    except ValueError:
        return {'error': f'Invalid IP address: "{ip_str}"'}

def print_results(result):
    """Print analysis results in a formatted way."""
    print("\n" + "="*50)
    print("📊 IP ANALYSIS RESULTS")
    print("="*50)
    
    if 'error' in result:
        print(f"❌ ERROR: {result['error']}")
        print("="*50 + "\n")
        return
    
    print(f"\n🌐 IP Address:     {result['ip']}")
    print(f"📌 Version:        {result['version']}")
    
    if result['version'] == 'IPv4':
        print(f"🔢 Class:          {result['class']}")
        print(f"🎯 Scope:          {result['scope']}")
        print(f"🔧 Default Subnet: {result['default_subnet']}")
        print(f"🎭 Subnet Mask:    {result['subnet_mask']}")
    else:
        print(f"📝 Type:           {result['type']}")
        print(f"🎯 Scope:          {result['scope']}")
        print(f"🔧 Default Subnet: {result['default_subnet']}")
    
    print("\n" + "="*50 + "\n")

# ============================================================
# MAIN PROGRAM - Enter IP address below
# ============================================================

print("🔍 IP Address Analyzer")
print("-" * 40)

# Method 1: Enter IP address manually
ip_input = input("Enter IP address (IPv4 or IPv6): ")
result = analyze_ip(ip_input)
print_results(result)

# Method 2: Test with predefined examples (uncomment to use)
"""
print("\n" + "📋 Testing with examples:")
print("-" * 40)

test_ips = [
    "8.8.8.8",        # Public IPv4
    "192.168.1.1",    # Private IPv4
    "127.0.0.1",      # Loopback
    "2001:db8::1",    # IPv6 Example
    "fe80::1",        # Link-Local IPv6
    "::1",            # IPv6 Loopback
]

for ip in test_ips:
    print(f"\n▶️  {ip}")
    result = analyze_ip(ip)
    if 'error' in result:
        print(f"   ❌ {result['error']}")
    else:
        if result['version'] == 'IPv4':
            print(f"   Class: {result['class']} | Scope: {result['scope']} | Subnet: {result['default_subnet']}")
        else:
            print(f"   Type: {result['type']} | Scope: {result['scope']} | Subnet: {result['default_subnet']}")
"""
```
